```python
# Building Request URL for sending a resquest to the CDC API
api_endpoint = 'https://data.cdc.gov/resource/'

def building_CDC_URL():
    rout_tojsonfile = 'vbim-akqf.json'
    CDC_URL = api_endpoint + rout_tojsonfile
    return CDC_URL 
```

```python
def making_APIreq(CDC_URL):
    resp = requests.get(CDC_URL)
    print(resp.status_code)
    if(int(resp.status_code) == 200):
        payload = json.loads(resp.text)
        return payload
    else:
        return{}
```

```python
result = making_APIreq(building_CDC_URL()) 
print(result)
```
